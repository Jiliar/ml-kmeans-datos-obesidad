#### **‚≠ïÔ∏è Variables con datos faltantes, Porcentaje total de valores faltantes en la base de datos.**

```python
import json

def valores_faltantes_por_columna(df):
    # Escribir las variables que tienen datos faltantes, adem√°s, proporcionar el porcentaje total de valores faltantes en la base de datos.
    # Paso 1: Encontrar columnas con valores faltantes
    faltantes_por_columna = df.isnull().sum()
    
    # Paso 2: Filtrar columnas que tienen valores faltantes
    data1 = faltantes_por_columna[faltantes_por_columna > 0]
    variables_con_faltantes = data1.to_dict()
    
    print('\nCant. valores faltantes por columna:',
          json.dumps(variables_con_faltantes, indent=4),
          sep="\n")
    
    # Paso 3: Calcular el porcentaje de valores faltantes por columna
    data = (data1 / len(df)) * 100
    data = data.to_dict()
    porcentaje_faltantes_por_columna = {key: f'{value:.2f}%' for key, value in data.items()}
    
    # Paso 4: Calcular el porcentaje total de valores faltantes en el DataFrame
    porcentaje_total_faltantes = (df.isnull().sum().sum() / df.size) * 100
    porcentaje_total_no_faltantes = (df.notnull().sum().sum() / df.size) * 100
    
    # Resultados
    print('\n% valores faltantes por columna:', f'{json.dumps(porcentaje_faltantes_por_columna, indent=4)}', sep="\n")
    
    print(f"\nPorcentaje total de valores faltantes en la base de datos: {porcentaje_total_faltantes:.2f}%")
    print(f"Porcentaje total de valores en la base de datos: {porcentaje_total_no_faltantes:.2f}%")

valores_faltantes_por_columna(df)
```

#### **‚≠ïÔ∏è Caracterizaci√≥n de Clasificaci√≥n de nivel de obesidad del individuo**

Al asignar un valor num√©rico a cada clase de obesidad, transformamos una variable cualitativa en un formato cuantitativo, lo que permite al algoritmo identificar patrones y similitudes entre los registros de manera m√°s eficiente. Esta transformaci√≥n facilita el c√°lculo de la proximidad entre individuos y mejora la precisi√≥n en la asignaci√≥n a grupos, apoyando un an√°lisis m√°s robusto y coherente de los niveles de obesidad en funci√≥n de las caracter√≠sticas cuantitativas.

### **‚≠ïÔ∏è Definici√≥n de valor K**

Es posible determinar el valor √≥ptimo de K utilizando una gr√°fica que represente la relaci√≥n entre el n√∫mero de clusters y la m√©trica de evaluaci√≥n del modelo, como el score de inercia. Al trazar esta gr√°fica, se busca identificar el "punto de codo", que es el lugar donde la tasa de mejora en el score comienza a disminuir significativamente. Este punto indica que agregar m√°s clusters no aporta un valor adicional considerable al modelo, lo que sugiere que el n√∫mero de clusters en ese momento es adecuado para la segmentaci√≥n de los datos.

En este caso, es viable utilizar StandardScaler en este caso. Los promedios de las variables (`FCVC, NCP, CH2O, FAF, TUE`) presentan diferentes rangos, lo que sugiere que tienen diferentes escalas. StandardScaler es adecuado para este tipo de situaciones, ya que convierte los datos a una distribuci√≥n con media `0` y desviaci√≥n est√°ndar `1`. Esto facilita que algoritmos como K-Means (que dependen de la distancia entre puntos) funcionen mejor, eliminando el efecto de las diferencias en escala.

```python
# Definir un rango de posibles valores para el n√∫mero de clusters a probar
Nc = range(1, 20)  # Evaluar desde 1 hasta 19 clusters

# Normalizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fijar una semilla para la reproducibilidad
np.random.seed(42)

# Crear una lista de modelos KMeans para cada n√∫mero de clusters en Nc
kmeans = [KMeans(n_clusters=i) for i in Nc]  # Inicializar el modelo KMeans para cada valor de Nc

# Calcular la puntuaci√≥n (score) para cada modelo KMeans ajustado a los datos X
score = [kmeans[i].fit(X_scaled).score(X_scaled) for i in range(len(kmeans))]  # Ajustar y obtener la puntuaci√≥n de cada modelo

# Configurar la figura para la visualizaci√≥n de la curva del codo
plt.figure(figsize=(8, 4))  # Establecer el tama√±o de la figura para la gr√°fica

# Graficar el n√∫mero de clusters versus la puntuaci√≥n obtenida
plt.plot(Nc, score, color='red')  # Trazar la curva del codo con l√≠nea roja

# A√±adir cuadr√≠cula de fondo en color gris claro
plt.grid(color='lightgray', linestyle='--', linewidth=0.7)  # Establecer la cuadr√≠cula con color y estilo

# Configurar el eje X para que muestre ticks de 3 en 3
plt.xticks(np.arange(1, 20, 1))  # Definir ticks en el eje X desde 1 hasta 19 en pasos de 3

# Etiquetas y t√≠tulo para el gr√°fico
plt.xlabel('Number of Clusters')  # Etiqueta del eje X que indica el n√∫mero de clusters
plt.ylabel('Score')  # Etiqueta del eje Y que indica la puntuaci√≥n del modelo
plt.title('Elbow Curve')  # T√≠tulo de la gr√°fica que describe su contenido

plt.show()  # Mostrar la gr√°fica generada en pantalla
```

La curva presenta una transici√≥n gradual, lo que indica que el rendimiento del modelo no muestra cambios dr√°sticos al aumentar el n√∫mero de clusters. En este contexto, se sugiere que 3 es un valor √≥ptimo para K, ya que proporciona un equilibrio entre la simplicidad del modelo y la capacidad de capturar la variabilidad de los datos.

### **‚≠ïÔ∏èComparaci√≥n de metodos de Clustering**

### *K-Means, Clustering Jer√°rquico, DBSCAN, Gaussian Mixture*

La comparaci√≥n de m√©todos de clustering en una soluci√≥n de machine learning no supervisada es fundamental para seleccionar el modelo que mejor se adapte a la estructura de los datos y a los objetivos del an√°lisis. Cada m√©todo, como K-Means, Clustering Jer√°rquico, DBSCAN y Gaussian Mixture, ofrece ventajas distintas: K-Means es eficiente y trabaja bien con datos esf√©ricos, mientras que el Clustering Jer√°rquico permite visualizar relaciones jer√°rquicas; DBSCAN es ideal para detectar cl√∫steres de formas arbitrarias y manejar ruido, y Gaussian Mixture se adapta a distribuciones el√≠pticas complejas. Evaluar estos enfoques permite identificar el m√©todo con mejor ajuste y precisi√≥n en la segmentaci√≥n, optimizando as√≠ los resultados en tareas como segmentaci√≥n de clientes, an√°lisis de patrones y detecci√≥n de anomal√≠as.

```python
k = 3

# Clustering K-Means
kmeans = KMeans(n_clusters=k, random_state=42)
labels_kmeans = kmeans.fit_predict(X_scaled)
silhouette_kmeans = silhouette_score(X_scaled, labels_kmeans)
print(f"Coeficiente de Silueta para K-Means: {silhouette_kmeans:.4f}")

# Clustering Jer√°rquico
agg = AgglomerativeClustering(n_clusters=k, linkage='ward')
labels_agg = agg.fit_predict(X_scaled)
silhouette_agg = silhouette_score(X_scaled, labels_agg)
print(f"Coeficiente de Silueta para Clustering Jer√°rquico: {silhouette_agg:.4f}")

# DBSCAN
silhouette_dbscan = 0
dbscan = DBSCAN(eps=2, min_samples=5)
labels_dbscan = dbscan.fit_predict(X_scaled)
mask = labels_dbscan != -1
if len(set(labels_dbscan)) > 1:
    silhouette_dbscan = silhouette_score(X_scaled[mask], labels_dbscan[mask])
    print(f"Coeficiente de Silueta para DBSCAN: {silhouette_dbscan:.4f}")
else:
    print("DBSCAN no encontr√≥ cl√∫steres distintos.")

# Modelo de Mezcla Gaussiana
gmm = GaussianMixture(n_components=k, covariance_type='full', random_state=42)
labels_gmm = gmm.fit_predict(X_scaled)
silhouette_gmm = silhouette_score(X_scaled, labels_gmm)
print(f"Coeficiente de Silueta para GMM: {silhouette_gmm:.4f}")
```

```python
# Ajustar el tama√±o de la figura
plt.figure(figsize=(20, 4))

# Crear la subgr√°fica
plt.subplot(1, 4, 1)

# Crear el gr√°fico de l√≠neas con seaborn
sns.lineplot(x=['K-Means', 'Jer√°rquico', 'DBSCAN', 'GMM'], y=[silhouette_kmeans, silhouette_agg, silhouette_dbscan, silhouette_gmm])

# T√≠tulo del gr√°fico
plt.title('Coeficiente de Silueta')

# Agregar cuadr√≠cula en gris claro
plt.grid(color='lightgray', linestyle='-', linewidth=0.5)

# Ajustar dise√±o para evitar superposiciones
plt.tight_layout()

# Mostrar la gr√°fica
plt.show()
```

```python
# Se extraen los centros de los clusters calculados por el modelo KMeans.
# Los centroides representan las posiciones medias de los grupos formados por el algoritmo.
C = kmeans.cluster_centers_ #centroides 

# Se imprime la matriz de centroides en la consola.
# Cada fila de esta matriz corresponde a un centro de cluster, y cada columna a una caracter√≠stica.
print(C)
```

```python
# Definir colores para representar las diferentes categor√≠as de clusters.
colores = ['green', 'purple', 'red']
asignar = []  # Lista para asignar colores a cada punto seg√∫n su cluster

# Iterar sobre las etiquetas de los clusters y asignar colores a cada punto.
for row in labels_kmeans:
    asignar.append(colores[row])

# Imprimir el n√∫mero total de puntos en la lista 'asignar' para verificar la longitud.
print(len(asignar))

# Crear una figura y un eje para el gr√°fico 3D.
fig = plt.figure(figsize=(8, 10))  # Especificar el tama√±o de la figura
ax = fig.add_subplot(111, projection='3d')  # Crear un subplot 3D

# Graficar los puntos en el espacio 3D, colore√°ndolos seg√∫n su cluster.
ax.scatter(X_scaled[:, 0], X_scaled[:, 1], X_scaled[:, 2], c=asignar, s=15, alpha=0.5)  # Tama√±o de los puntos reducido a 15

# Graficar los centros de los clusters, utilizando un marcador diferente ('*') para destacarlos.
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c=colores, s=100, alpha=1)  # Tama√±o de las estrellas reducido a 100

# Trazar una l√≠nea negra entre los centros de los clusters
ax.plot(C[:, 0], C[:, 1], C[:, 2], color='black', linestyle='-', linewidth=0.5)  # L√≠nea negra que conecta los centros para visualizar los centros

# Etiquetar los ejes y establecer un t√≠tulo para el gr√°fico.
ax.set_xlabel('Eje X', fontsize=8)
ax.set_ylabel('Eje Y', fontsize=8)
ax.set_zlabel('Eje Z', fontsize=8)
ax.set_title('Gr√°fica 3D de Cl√∫steres', fontsize=10)

# Mostrar el gr√°fico resultante en una ventana.
plt.show()
```
